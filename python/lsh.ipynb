```python
import random
```

```python
# universe
U = list(range(1,1000))

# two similar sets
C1 = list(range(1,700))
C2 = list(range(100,1000))
# Jaccard sim = 60%

# generate a random permutation of U
random.shuffle(U) # new U is a permutation of the previous U

def minhash(C, U):
  for x in U:
    if x in C: 
      return x


m = 100
agreements = 0;
for i in range(m):
  random.shuffle(U)
  minhashC1 = minhash(C1,U)
  minhashC2 = minhash(C2,U)
  if minhashC1 == minhashC2:
    agreements +=1 
  print(minhashC1, minhashC2)

print('Jaccard similarity estimation', agreements/m)
```
